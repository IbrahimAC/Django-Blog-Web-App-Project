##**STEP 5)** Automatically resizing images

###**Overview:**

We want to have our images are automatically resized when we upload them to prevent extremely large images on our file system as we are only displaying a small image on our Profiles page.

**Problem: **Having images that are e.g. 4000 pixels big just for it to be scaled down to 125 pixels. It takes up alot of space on the file system and causes the website to run slow as it has to send that large image to  the browser everytime.

To resize images when we upload them we are going to use a package called pillow.

We previously installed this when we first used the imagefield in the previous lab.

**A)** In our users app, open our models.py file. Implement this method WITHIN our 'Profile' class.

```python
def save(self):
  super().save() #This is how we access the parents classes save method using super()

  
```

This 'save' method gets run after our model is saved, it alredy exists in our parent class however we are making our own to add more functionality.

**What is super()?**

The Python super() method lets you access methods from a parent class from within a child class.

**B)** To grab the image that is saved and resize it we need the import that Pillow library.

Implement the following import, into the same file.



```python
from PIL import Image
```

**C)** We need to open the image for the profile instance, to do this we implement the following:

```python
def save(self):
  super().save()
  #Add the code line below to your function.
  img = Image.open(self.image.path) #This will open the image of the current instance
```

Pick an amount to resize to, something similar to what we display on the front end of our website (125 pixels).

Around 300 pixels is fine.  

Any images larger than whatever you specify, will be sized down.

**D)** Add this code under the recently previously implemented line of code. 

This will check if an image's height or width excedes e.g. 300 pixels, and if so it will be resized to whatever the height and width is in the 'output_size' variable

Once resized, we simply save it back to the same path which should overwrite the original image.

**Note:** I am using 300 pixels but feel free to change that slightly.

```python
 if img.height > 300 or img.width >300:
    output_size = (300, 300) #What H and W we want to resize it too
    img.thumbnail(output_size) #Resizing the image
    img.save(self.image.path) #Saving the now resized image to the same path to overwrite it
```

**Your 'save' method should end up looking like this.**

```python
def save(self):
  super().save()
.
  img = Image.open(self.image.path)

  if img.height > 300 or img.width >300:
    output_size = (300, 300) 
    img.thumbnail(output_size)
    img.save(self.image.path)  
```
