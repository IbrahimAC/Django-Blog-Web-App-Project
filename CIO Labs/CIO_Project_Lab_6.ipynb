## Step 2 

**Overview:** Now we are going to add a table and fields using Django ORM and models. 

**A)** Inside your blog/models.py copy and paste the following code: 

```python
from django.db import models
from django.utils import timezone
from django.contrib.auth.models import User


class Post(models.Model):
    title = models.CharField(max_length=100)
    content = models.TextField()
    date_posted = models.DateTimeField(default=timezone.now)
    author = models.ForeignKey(User, on_delete=models.CASCADE)

    def __str__(self):
        return self.title
```

**F)** Now lets push our SQL code to our SQL-Lite database using the following terminal command. 

```python
python manage.py migrate 
```

**Conclusion:** Using Django ORM's and models we have been able to create a full SQL table stored in a SQLlite database without having to learn or write all of these lines of SQL code. Migrations allow us to update and change our SQL table while data is stored inside it, without having to write many lines of SQL code. Very neat right?
