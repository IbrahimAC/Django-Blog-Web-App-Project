##**Step 2)** Understanding Paginator Objects

###Overview: 

We want to only display a certain amount of posts per page and include a link to the next page too.

Solution: Django makes this easy for us, as we can use something called a 'paginator object'

**A)** Stop your server and reopen your python shell, so we can import the Paginator class.

Excecute each line of code on it's own

```python
python manage.py shell #If you haven't already opened your shell

from django.core.paginator import Paginator
#Now we need to specify which posts we want to paginate
#We will just specify a couple for now
posts = ['1', '2', '3', '4', '5'] #Assume this is a list of articles
#Remain in your shell when reading step B
```

**B)** Now we need to create a paginator object of these posts, we will make it so it is 2 posts per page

In your shell input the following

```python
p = Paginator(posts, 2)
#We have 5 posts being paginated, with 2 per page meaning we should have 3 pages
p.num_pages #will return how many pages we have
#stay in your python shell
```

If we want to loop over these we could use the 'page_range method', this will a Python range we can loop over.

**C)** In your shell input the code below

```python
for page in p.page_range: #this is an attribute not a method so no '()' needed
  print(page)
```

It should print the pages

```python
#Output from the code cell above
1
2
3
```

**D)** We can access it by writing

```python
p1 =p.page(1)

p1

#Output of line 3
<Page 1 of 3> #We can see it's page 1 of 3

#To get the specific integer for the page, write this
p1.number
#Output
1

p1.object_list #Returns the amount of objects on that page
#Output
['1', '2']
```

The most useful thing we  nee for when we create links for our pages is whether or not they have previous pages or pages after it.

Page 1 shouldn't have any previous pages we can check that with

```python
p1.has_previous()
#Output
False #There are no previous pages

#If we check if it has a page after it we would use the following
p1.has_next()
#Output
True #There is a page after it

#To get the page number of the next page it is
p1.next_page_number()
#Output
2 #The next page.

exit()
```

We should now be familiar some of the functionality of Paginator Objects. This is important as we will want to use these on our site.

```python
class UserPostListView(ListView):
    model = Post
    template_name = 'blog/user_posts.html'  # <app>/<model>_<viewtype>.html
    context_object_name = 'posts'
    paginate_by = 5
```

This will still display all the posts from our post model but we want to add a filter which only displays the users post.

We do this by creating a new URL pattern.

First to override the query that the listview returns (all the posts) we need to override the method known as 'get_queryset' and change it within that.

**B)** Before we implement the function, in order for it to work we need to first import some things.

**In your django.shortcuts import add the following:**

get_object_or_404

This fucntion will attempt to get an object from the database however if it does not exist it will return a 404.

**We also need to import our user model, import this:**

from django.contrib.auth.models import User

Those imports  should look similar to the code below now

```python
from django.shortcuts import render, get_object_or_404
from django.contrib.auth.models import User
```

**C)** Now within your UserPostListView class implement the following.

```python

    def get_queryset(self):
        user = get_object_or_404(User, username=self.kwargs.get('username'))
        return Post.objects.filter(author=user).order_by('-date_posted')
```

We are getting to User, and then the username from the URL. If that user exists it will be used in the 'User' variable, if it doesn't a 404 will be returned.

In the return is where we filter the posts, only allowing posts where the specific User is the author of to show, in newest to oldest order.
